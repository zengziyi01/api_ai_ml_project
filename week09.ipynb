# 基于LBS环境下的 高德POI

* 搜索服务API是一类简单的HTTP接口，提供多种查询POI信息的能力，其中包括关键字搜索、周边搜索、多边形搜索、ID查询四种筛选机制

> 1. 关键字搜索：通过用POI的关键字进行条件搜索，例如：肯德基、朝阳公园等；同时支持设置POI类型搜索，例如：银行
> 2. 周边搜索：在用户传入经纬度坐标点附近，在设定的范围内，按照关键字或POI类型搜索；
> 3. 多边形搜索：在多边形区域内进行搜索
> 4. ID查询：通过POI ID，查询某个POI详情，建议可同输入提示API配合使用

* 准备url接口
> https://restapi.amap.com/v3/place/text?parameters

* 查询方式
> 1. keywords(查询关键字)
> 2. types(查询POI类型)

## 基于搜索关键词

* 准备KEY  
> 7e2576117dece0a92adda7880d77130d

```python
import requests
```

```python
POI_url = 'https://restapi.amap.com/v3/place/text?parameters'

payload = {
    'key':'7e2576117dece0a92adda7880d77130d',
    'keywords':'广州南方学院',
    'extensions':'all',    
}

r = requests.get(url=POI_url, params=payload )
```

## 基于周边搜索

```python
import requests
geo_url = "https://restapi.amap.com/v3/geocode/geo?parameters"
payload = {
    "key":"7e2576117dece0a92adda7880d77130d",
    "address":"广州南方学院"
}
r = requests.get(url=geo_url, params=payload)
广州南方学院_location = r.json()['geocodes'][0]['location']
广州南方学院_location
```

```python
around_url = 'https://restapi.amap.com/v3/place/around?parameters'

payload = {
    "key":"7e2576117dece0a92adda7880d77130d",
    "location":广州南方学院_location,
    "keywords":"中餐厅"
}
r = requests.get(around_url,params=payload)
```

```python
r.json()
```

* 使用编码表方式

```python
around_url = 'https://restapi.amap.com/v3/place/around?parameters'

payload = {
    "key":"7e2576117dece0a92adda7880d77130d",
    "location":广州南方学院_location,
    "types":"050100",
    'page':3
}
r = requests.get(around_url,params=payload)
```
